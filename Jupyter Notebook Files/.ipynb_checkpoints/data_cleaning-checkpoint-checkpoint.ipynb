# Data Cleaning

```python
mvps = pd.read_csv("mvps.csv")
```

```python
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
```

```python
mvps[mvps['Year'] == 2022]
```

```python
players = pd.read_csv("players.csv")
```

```python
players
```

```python
del players["Unnamed: 0"]
del players["Rk"]
```

```python
players.head()
```

```python
players["Player"].head(50)
```

```python
players['Player'] = players["Player"].str.replace("*", "", regex = False)
```

```python
players.head(5)
```

```python
players.groupby(['Player', 'Year'])
```

```python
def single_row(df):
    # if there is only one row, don't do processing
    if df.shape[0] == 1:
        return df
    else:
        row = df[df['Tm'] == "TOT"]
        # if this person played for multiple teams, replace them with the last team they played with
        row["Tm"] = df.iloc[-1,:]["Tm"]

# Making a group for each player/year combination and for this group, only keep one unique row (with the last team they played for)
players = players.groupby(["Player", "Year"]).apply(single_row)        
```

```python
players.head(20)
```

```python
players.index = players.index.droplevel()
players.index = players.index.droplevel()
```

```python
players[players["Player"] == "Greg Anderson"]
```

```python
# outer merge since mvps table doesn't contain info about all players, and we want to keep this information
combined = players.merge(mvps, how = "outer", on=["Player", "Year"])
```

```python
combined.head()
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined
```

```python
teams = pd.read_csv("teams.csv")
```

```python
teams
```

```python
teams.head(30)
```

```python
# ~ is the NOT operator
teams = teams[~teams["W"].str.contains("Division")]
```

```python
teams.head(30)
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex = False)
```

```python
teams.head(5)
```

```python
teams["Team"].unique()
```

```python
combined["Tm"].unique()
```

```python
nicknames = {}

with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        # split this by comma 
        abbrev,name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined.head(5)
```

```python
stats = combined.merge(teams, how = "outer", on = ["Team", "Year"])
```

```python
stats
```

```python
combined
```

```python
del stats["Unnamed: 0"]
```

```python
# if error, ignore and just don't convert - i.e., there are some strings
stats = stats.apply(pd.to_numeric, errors = "ignore")
```

```python
stats.dtypes
```

```python
stats["GB"].unique()
```

```python
stats["GB"] = stats['GB'].str.replace("—", "0")
```

```python
stats["GB"].unique()
```

```python
stats['GB'] = pd.to_numeric(stats["GB"])
```

```python
stats.dtypes
```

```python
stats.to_csv("player_mvp_stats.csv")
```

```python
stats = pd.read_csv('player_m')
```

```python
# anyone who played more than 70 games, take top 10
highest_scoring = stats[stats["G"] > 70].sort_values("PTS", ascending=False).head(10)
```
